- Scaling(척도) : 각 데이터 특성들에 대해 다음과 같은 선형 변환을 통해 전체 특성들의 통계적 분포를 동일하게 맞추는 과정

 - StandardScaler
   - (0 ,0)을 중심으로 데이터가 분포하게 된다. 
   - 각각의 특성값에 대해서 평균을 빼고 표준편차(얼마나 퍼져있는지)로 나눠 준다. 
   - outlier(이상치)에 영향을 많이 받는다.
   - 각 특성들의 값이 차이가 많이 크더라도 같은 스케일 내에 데이터가 존재하게 하기 때문에 학습이 가능하게 만들어줌 
  - MinMaxScaler
   - 모든 특성이 정확하게 0과 1사이에 위치하도록 데이터를 변경
   - oulier에 영향을 받지 않는다.
   - 많이 쓰이는 scaler 이다.
  - RobustScaler
   - 중앙값을 이용해서 척도를 표준화하는 방법
   - median 값이 0이 되도록 함
   - 전체 특성 값과 매우 동떨어진 특성 값에 영향을 받지 않음.
   - 실제로 책에서도 많이 사용되지는 않음.
  - Normalizer(정규화)
   - 크기가 1짜리인 원을 그리고 그 원의 원주에 데이터가 mapping 되는 형식이다.
   - 크기의 데이터가 없어진다. 각도(방향) 데이터만 남게 된다.
   - 특성 벡터의 길이는 무관하고 데이터의 방향만이 중요할 때 많이 사용하게 된다.
   - 종류
     - l1
     - l2(기본)
     - max
